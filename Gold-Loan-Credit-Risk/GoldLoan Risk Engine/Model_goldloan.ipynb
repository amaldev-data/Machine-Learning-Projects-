# Visualisation

```python
plt.figure(figsize=(10,5))

ax = sns.countplot(
    x="gender",
    data=df,
    palette="viridis",
    width=0.2  # reduce bar width
)

# Add count labels on top of bars
for bar in ax.patches:
    height = bar.get_height()
    ax.annotate(
        f'{int(height)}',
        (bar.get_x() + bar.get_width() / 2, height),
        ha='center',
        va='bottom',
        fontsize=10,
        xytext=(0, 3),
        textcoords='offset points'
    )

plt.title("Gender Distribution")
plt.xlabel("Gender")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()
```

```python
# Create figure with 2 subplots
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# -------------------------------
# 1. Gender Distribution
# -------------------------------
ax1 = sns.countplot(
    x="gender",
    data=df,
    palette="viridis",
    width=0.4,
    ax=axes[0]
)

# Add count labels
for bar in ax1.patches:
    height = bar.get_height()
    ax1.annotate(
        f'{int(height)}',
        (bar.get_x() + bar.get_width()/2, height),
        ha='center',
        va='bottom',
        fontsize=10
    )

axes[0].set_title("Gender Distribution")
axes[0].set_xlabel("Gender")
axes[0].set_ylabel("Count")


# -------------------------------
# 2. Gender-wise Default Percentage
# -------------------------------

# Calculate percentage
gender_default = (
    df.groupby("gender")["default_status"]
    .mean() * 100
).reset_index()

ax2 = sns.barplot(
    x="gender",
    y="default_status",
    data=gender_default,
    palette="magma",
    width=0.4,
    ax=axes[1]
)

# Add percentage labels
for bar in ax2.patches:
    height = bar.get_height()
    ax2.annotate(
        f'{height:.2f}%',
        (bar.get_x() + bar.get_width()/2, height),
        ha='center',
        va='bottom',
        fontsize=10
    )

axes[1].set_title("Gender-wise Default Percentage")
axes[1].set_xlabel("Gender")
axes[1].set_ylabel("Default %")

plt.tight_layout()
plt.show()
```

```python


# Create subplots
fig, axes = plt.subplots(1, 2, figsize=(16, 5))

# -----------------------------
# Left Plot - State Distribution
# -----------------------------
ax1 = sns.countplot(
    x="state",
    data=df,
    palette="viridis",
    width=0.5,
    ax=axes[0]
)

# Add count labels
for bar in ax1.patches:
    height = bar.get_height()
    ax1.annotate(
        f'{int(height)}',
        (bar.get_x() + bar.get_width() / 2, height),
        ha='center',
        va='bottom',
        fontsize=9,
        xytext=(0, 3),
        textcoords='offset points'
    )

axes[0].set_title("State Distribution")
axes[0].set_xlabel("State")
axes[0].set_ylabel("Frequency")
axes[0].tick_params(axis='x', rotation=45)


# -----------------------------
# Right Plot - State by Default Status
# -----------------------------
ax2 = sns.countplot(
    x="state",
    hue="default_status",
    data=df,
    palette="viridis",
    ax=axes[1]
)

# Add count labels
for bar in ax2.patches:
    height = bar.get_height()

    if height > 0:
        ax2.annotate(
            f'{int(height)}',
            (bar.get_x() + bar.get_width() / 2, height),
            ha='center',
            va='bottom',
            fontsize=8,
            xytext=(0, 3),
            textcoords='offset points'
        )

axes[1].set_title("State by Default Status")
axes[1].set_xlabel("State")
axes[1].set_ylabel("Count")
axes[1].tick_params(axis='x', rotation=45)


# Custom legend labels
handles, labels = ax2.get_legend_handles_labels()
ax2.legend(handles, ["No Default", "Default"], title="Default Status")

plt.tight_layout()
plt.show()
```

```python
# Total customers by state
state_total = df["state"].value_counts()

# Total defaults by state
state_default = (
    df[df["default_status"] == 1]
    .groupby("state")
    .size()
)

# Default percentage by state
default_percentage = (
    (state_default / state_total) * 100
).sort_values(ascending=False)

# Plot
plt.figure(figsize=(10,5))

ax = sns.barplot(
    x=default_percentage.index,
    y=default_percentage.values,
    palette="viridis"
)

# Add percentage labels
for i, value in enumerate(default_percentage.values):
    ax.text(
        i,
        value + 0.3,
        f"{value:.1f}%",
        ha='center',
        fontsize=9
    )

plt.title("Default Percentage by State")
plt.xlabel("State")
plt.ylabel("Default Percentage (%)")
plt.xticks(rotation=45)


plt.tight_layout()
plt.show()
```

```python
from matplotlib.patches import Patch

# Create subplots
fig, axes = plt.subplots(1, 2, figsize=(16, 5))

# -----------------------------
# Left Plot - Histogram
# -----------------------------
sns.histplot(
    data=df,
    x='age',
    bins=20,
    kde=True,
    color='purple',
    ax=axes[0]
)

axes[0].set_xticks(range(20, 71, 5))
axes[0].set_title("Age Distribution")
axes[0].set_xlabel("Age")
axes[0].set_ylabel("Count")
axes[0].grid(True, linestyle='--', alpha=0.5)

# -----------------------------
# Right Plot - KDE by Default Status
# -----------------------------
sns.kdeplot(
    data=df,
    x="age",
    hue="default_status",
    fill=True,
    common_norm=False,
    alpha=0.5,
    ax=axes[1]
)

axes[1].set_xticks(range(20, 71, 5))
axes[1].set_title("Age Distribution by Default Status")
axes[1].set_xlabel("Age")
axes[1].set_ylabel("Density")
axes[1].grid(True, linestyle='--', alpha=0.5)

# Remove seaborn automatic legend
if axes[1].get_legend() is not None:
    axes[1].get_legend().remove()

# Custom legend
legend_colors = ["#ff7f0e", "#1f77b4"]

legend_labels = ["No Default", "Default"]

handles = [
    Patch(facecolor=legend_colors[i], label=legend_labels[i], alpha=0.5)
    for i in range(2)
]

axes[1].legend(handles=handles, title="Default Status")

plt.tight_layout()
plt.show()
```

```python

# Total occupation counts
occupation = df["occupation"].value_counts()

# Default counts by occupation
default_counts = (
    df[df["default_status"] == 1]
    .groupby("occupation")
    .size()
)

# Common occupation order
occupation_order = occupation.index

# Create consistent colors for occupations
colors = sns.color_palette("plasma", len(occupation_order))
color_map = dict(zip(occupation_order, colors))

# Create subplots
fig, axes = plt.subplots(1, 2, figsize=(16, 5))

# -----------------------------
# Total Occupation Count Plot
# -----------------------------
sns.barplot(
    x=occupation.index,
    y=occupation.values,
    palette=[color_map[i] for i in occupation.index],
    ax=axes[0]
)

# Labels on bars
for i, value in enumerate(occupation.values):
    axes[0].text(
        i,
        value + 20,
        str(value),
        ha='center',
        fontsize=9
    )

axes[0].set_title("Total Customers by Occupation")
axes[0].set_xlabel("Occupation")
axes[0].set_ylabel("Count")
axes[0].tick_params(axis='x', rotation=45)

# -----------------------------
# Default Count Plot
# -----------------------------
sns.barplot(
    x=occupation_order,
    y=default_counts.reindex(occupation_order).values,
    palette=[color_map[i] for i in occupation_order],
    ax=axes[1]
)

# Labels on bars
for i, value in enumerate(default_counts.reindex(occupation_order).values):
    axes[1].text(
        i,
        value + 10,
        str(int(value)),
        ha='center',
        fontsize=9
    )

axes[1].set_title("Total Defaults by Occupation")
axes[1].set_xlabel("Occupation")
axes[1].set_ylabel("Defaults")
axes[1].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```

# Model Training

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler 
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, classification_report, roc_auc_score, accuracy_score
import xgboost as XGBClassifier
```

```python
# Risk map is hardcoded by YOU, not learned from data
occupation_risk_map = {
    'Professional': 1,
    'Salaried':     2,
    'Small Business': 3,
    'Farmer':       4,
    'Daily Wage':   5
}
df['occupation_encoded'] = df['occupation'].map(occupation_risk_map)
```

```python
print(df['occupation_encoded'].value_counts())
print(df['occupation'].value_counts())
```

```python
df
```

```python
X = df[['occupation_encoded','past_defaults', 'active_loans', 'total_debt_burden', 'interest_rate_pct',
        'ltv_ratio', 'emi_inr', 'cibil_score', 'monthly_income',
        'emp_years', 'income_to_emi', 'loan_amount_inr', 'tenure_months']]

Y = df['default_status']
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42,stratify=Y)
```

```python
scaler=StandardScaler()
X_train_scaled = scaler.fit_transform(x_train)
X_test_scaled = scaler.transform(x_test)
```

# Logistic Regression

```python
lr = LogisticRegression(max_iter=1000, random_state=42, class_weight='balanced')
lr.fit(X_train_scaled, y_train)

y_pred_lr  = lr.predict(X_test_scaled)
y_proba_lr = lr.predict_proba(X_test_scaled)[:, 1]

print("=== Logistic Regression ===")
print(classification_report(y_test, y_pred_lr))
print("ROC-AUC:", roc_auc_score(y_test, y_proba_lr))
```

# XG Boost Classifier

```python
from xgboost import XGBClassifier
from sklearn.metrics import classification_report, roc_auc_score

neg, pos = y_train.value_counts()

xgb_model = XGBClassifier(
    n_estimators=100,
    learning_rate=0.1,
    max_depth=4,
    random_state=42,
    eval_metric='logloss',
    use_label_encoder=False,
    scale_pos_weight=neg/pos
)
xgb_model.fit(X_train_scaled, y_train)

y_pred_xgb  = xgb_model.predict(X_test_scaled)
y_proba_xgb = xgb_model.predict_proba(X_test_scaled)[:, 1]

print("=== XGBoost ===")
print(classification_report(y_test, y_pred_xgb))
print("ROC-AUC:", roc_auc_score(y_test, y_proba_xgb))

print()
```

```python
lr_acc= roc_auc_score(y_test, y_proba_lr)
print(f"Logistic Regression ROC-AUC :{lr_acc * 100:.2f} %")

xgb_acc = roc_auc_score(y_test, y_proba_xgb)
print(f"XGBoost ROC-AUC :{xgb_acc * 100:.2f} %")
```

```python
confusion_lr = confusion_matrix(y_test, y_pred_lr)
confusion_xgb = confusion_matrix(y_test, y_pred_xgb)    
plt.figure(figsize=(7,3))

plt.subplot(1, 2, 1)
sns.heatmap(confusion_lr, annot=True, fmt='d', cmap='Blues', cbar=False)
plt.title("Confusion Matrix - Logistic Regression") 

plt.subplot(1, 2, 2)
sns.heatmap(confusion_xgb, annot=True, fmt='d', cmap='Greens', cbar=False)
plt.title("Confusion Matrix - XGBoost") 

plt.tight_layout()
plt.show()
```
